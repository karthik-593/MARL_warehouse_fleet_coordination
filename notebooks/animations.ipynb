# Warehouse MARL — Simulation Animations

Seven interactive HTML animations covering all three training stages.
All navigation uses the **real trained PPO policy** — no BFS.

| # | Scenario |
|---|----------|
| 1 | Navigation — 5×5 plain grid |
| 2 | Navigation — 10×10 plain grid |
| 3 | Navigation — 10×10 + stationary obstacles |
| 4 | Navigation — 10×10 + moving obstacles |
| 5 | Navigation — L6 warehouse + 2 frozen robot obstacles |
| 6 | Stage 2 Assignment DQN — single robot (best of 20 seeds) |
| 7 | Stage 3 MAPPO — 3-robot coordination (best of 20 seeds) |

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

from IPython.display import HTML, display, Markdown
from utils.visualize import build_scenario_catalog, make_html_animation
from training.train_mappo import CKPT_DIR, DEVICE

OUT_DIR = os.path.join(CKPT_DIR, 'animations')
os.makedirs(OUT_DIR, exist_ok=True)

# FPS per scenario
FPS = {1: 10, 2: 8, 3: 8, 4: 8, 5: 8, 6: 5, 7: 4}

print(f'Device  : {DEVICE}')
print(f'Output  : {OUT_DIR}')

CAT = build_scenario_catalog(CKPT_DIR, DEVICE)

def run(n, force_regen=False):
    s    = CAT[n]
    path = os.path.join(OUT_DIR, s['out'])
    display(Markdown(f'### {n}. {s["title"]}'))
    display(Markdown(f'*{s["desc"]}*'))
    if force_regen and os.path.exists(path):
        os.remove(path)
    if not os.path.exists(path):
        frames = s['record']()
        make_html_animation(frames, path, fps=FPS[n])
    display(HTML(open(path, encoding='utf-8').read()))
```
